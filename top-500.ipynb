```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_full_df = pd.read_csv('data/GLC24_PA_metadata_train.csv')
train_full_df.speciesId = train_full_df.speciesId.astype(int)
```

```python
species = pd.DataFrame()
species['count'] = train_full_df['speciesId'].value_counts().sort_values(ascending=False)
species['speciesId'] = species.index
species = species.reset_index(drop=True)
```

```python
species.iloc[:500]
```

```python
species.plot(kind='line', y='count', figsize=(20, 10), xticks=range(0, 5200, 200))
```
